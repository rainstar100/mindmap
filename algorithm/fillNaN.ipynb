```python
import numpy as np
import pandas as pd
```

```python
mydata=pd.read_csv("info.csv",index_col=0)
mydata.head()
```

```python
#replace NaN with number,for example 0,100
mydata.fillna(100)

```

```python
#repalce NaN with the mean of data
#mydata.fillna(mydata.mean())

mydata.fillna(mydata['yuwen'].mean())
```

```python
mydata
```

```python
#replace NaN with the previous value
#mydata.fillna(method='ffill')
mydata.fillna(method='ffill',axis=1)
```

```python
#replace NaN with the predic value by neural net
from sklearn import neural_network
train_x=mydata.iloc[4:9,0:2].values
#train_x
train_y=mydata.iloc[4:9,2].values
#train_y
model=neural_network.MLPClassifier(max_iter=1000)
model.fit(train_x,train_y)
predict_x=mydata.iloc[1:3,0:2]
predict_x
predict_y=model.predict(predict_x)
predict_y

mydata.loc[mydata['shuxue'].isnull(),'shuxue']=predict_y.tolist()
mydata
```
